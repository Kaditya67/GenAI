## Advanced Operation with Business Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Data of some restaurants in a city
# DS : [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 1000, 1200, 1300, 1400],
    [2, 800, 900, 950, 1000],   
    [3, 1500, 1600, 1700, 1800],
    [4, 700, 750, 800, 850],
    [5, 1200, 1300, 1400, 1500],
    [6, 600, 650, 700, 750],
    [7, 1100, 1200, 1300, 1400],
    [8, 500, 550, 600, 650],
    [9, 900, 950, 1000, 1050],  
    [10, 1300, 1400, 1500, 1600]
])

print("==================== Sales Data Analysis ====================")
# print("Sales Data:\n", sales_data)
print("Shape of Sales Data:", sales_data.shape)
# [3] --> Third Record
# [:3] --> First Three Records
# [:, 1] --> All Rows but only Year 2021
# [:, 1:] --> All Rows, all columns except Restaurant ID
print("Sample ddata for 1st three restaurants:\n", sales_data[:, 1])
```

#### Total Sales

```python
print("Sum of Sales in 2021:", np.sum(sales_data[:, 1]))
print("Sum of Sales for all years:\n", np.sum(sales_data[:, 1:], axis=0))   # Column-wise sum
print("Sum of Sales for all restaurants:\n", np.sum(sales_data[:, 1:], axis=1))  # Row-wise sum
```

#### Min and Max Sales

```python
min_sales = np.min(sales_data[:, 1:], axis=0)  # Min sales for each year
print("Minimum Sales for each year:", min_sales)

max_sales = np.max(sales_data[:, 1:], axis=0)  # Max sales for each year
print("Maximum Sales for each year:", max_sales)

diff_sales = max_sales - min_sales  # Difference between max and min sales for each year
print("Difference between Max and Min Sales for each year:", diff_sales)

# Print Index at which max sales occurred
max_sales_index = np.argmax(sales_data[:, 1:], axis=0)  # Index of max sales for each year
print("Index of Max Sales for each year:", max_sales_index)

# Cumulative Sales
cumulative_sales = np.cumsum(sales_data[:, 1:], axis=1)  # Cumulative sum of sales for each year'
print("Cumulative Sales for each year:\n", cumulative_sales)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumulative_sales, axis=0))
plt.title('Average Sales Over Time')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.grid(True)
plt.legend()
plt.show()
```

### Vector

```python
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])
print("Vector Addition:", vector1 + vector2)
print("Vector Subtraction:", vector1 - vector2)
print("Vector Multiplication:", vector1 * vector2) 
print("Vector Scalar Multiplication:", vector1 * 2)  # Scalar multiplication
print("Vector Division:", vector1 / vector2)

print("Dot Product:", np.dot(vector1, vector2))  # Dot product
print("Cross Product:", np.cross(vector1, vector2))  # Cross product
```

$$
\cos(\theta) = \frac{A \cdot B}{\|A\| \cdot \|B\|}
$$


```python
cos_theta = np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))
theta_rad = np.arccos(cos_theta)
theta_deg = np.degrees(theta_rad)

print(theta_deg)  # 90.0 degrees
```
