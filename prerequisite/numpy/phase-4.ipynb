## Image Conversion using Numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.random.rand(3,3)
arr3 = np.zeros((4,4))
```

```python
np.save('array1.npy', arr1)
np.save('array2.npy', arr2)
np.save('array3.npy', arr3)
```

```python
loaded_arr1 = np.load('array1.npy')
print("Loaded Array 1:\n", loaded_arr1)
```

```python
try:
    logo = np.load('numpy-logo.npy')
    print("Numpy Logo Loaded Successfully: ",logo)
    # Display the logo

    dark_logo = 1 - logo
    # logo = dark_logo 

    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(logo, cmap='gray')
    plt.title('Numpy Logo')
    plt.grid(False)
    plt.axis('off')

    plt.subplot(122)
    plt.imshow(dark_logo, cmap='gray')
    plt.title('Inverted Numpy Logo')
    plt.grid(False)
    plt.axis('off')

    plt.tight_layout()
except FileNotFoundError:
    print("File 'numpy-logo.npy' not found. Please ensure the file exists in the current directory.")
```
