### Understanding Pyplot

###### It keeps track of canvas until plt.show() is called

| Feature / Aspect              | **Pyplot Interface (`plt`)** | **Object-Oriented API (`fig`, `ax`)** |
|--------------------------------|------------------------------|----------------------------------------|
| **Purpose**                   | Quick, simple plotting (MATLAB-like) | Fine-grained control, complex layouts |
| **Ease of Use**               | Very easy for beginners      | Steeper learning curve                 |
| **Code Style**                 | Implicit — relies on *global state* (current figure/axes) | Explicit — you directly work with `Figure` & `Axes` objects |
| **Flexibility**               | Less flexible, limited for multi-plot layouts | Highly flexible, supports advanced customization |
| **Best Use Case**              | Simple, one-off plots, exploratory analysis | Large scripts, dashboards, multiple plots in one figure |
| **Figure & Axes Management**  | Automatic — Matplotlib decides where to draw | Manual — you create and pass around figure/axes objects |
| **Performance in Large Codebases** | Can become messy and hard to debug | Scales well, easier to maintain |


#### Pyplot functions

###### plt.plot(): Line Graph
###### plt.grid():  Grid
###### plt.xlabel() / plt.ylabel(): Label Axis
###### plt.title(): Title Canvas
###### plt.legend()
###### plt.xticks() / plt.yticks()
###### plt.xlim() / plt.ylim()
###### plt.show()

#### Line Chart

```python
# plt.plot(x,y, marker='o', color='blue', linestyle='--', linewidth=2, markersize=8, label='Data Points')

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [100, 150, 120, 250, 200]
plt.plot(months, sales, marker='o', color='blue', linestyle='--', linewidth=2, markersize=8, label='2025 Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales (in Hundreds)')
plt.title('2025 Sales Data')

# plt.legend()
plt.legend(loc='upper left', fontsize=12)        # loc = 'upper left' / 'upper right' / 'lower left' / 'lower right'

# plt.grid(True)
plt.grid(linestyle=':', linewidth=0.5, alpha=0.7, color='gray')  # linestyle, linewidth, alpha

plt.xlim(-1, 5)  # Set x-axis limits
plt.ylim(50, 300)  # Set y-axis limits

# Customizing ticks
plt.xticks([0, 1, 2, 3, 4, 5], ['M1', 'M2', 'M3', 'M4', 'M5', 'M6'], rotation=45)  
plt.yticks([0,50, 100, 150, 200, 250, 300], ['zero','50', '100', '150', '200', '250', '300'])

plt.show()
```

#### Bar Chart

```python
# plt.bar(x, height, color='skyblue', edgecolor='black', width=0.4, label='Bar Chart')
products = ['A', 'B', 'C', 'D']
sales = [150, 200, 300, 250]
# plt.barh(products, sales, color='skyblue', edgecolor='black', height=0.4, label='2025 Sales Data') # use height for horizontal bar chart
plt.bar(products, sales, color='skyblue', edgecolor='black', width=0.4, label='2025 Sales Data')  # use width for vertical bar chart
plt.xlabel('Products')
plt.ylabel('Sales (in Hundreds)')
plt.title('2025 Product Sales Data')
plt.legend()
# plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

#### Histogram Chart

```python
# plt.hist(values, bins=5, color='skyblue', edgecolor='black', alpha=0.7, rwidth=0.9)  # rwidth = relative width of bars

scores = [55, 65, 75, 85, 95, 100, 60, 70, 80, 90, 50, 40, 30, 20, 10]
plt.hist(scores, bins=5, color='skyblue', edgecolor='black', alpha=0.5, rwidth=0.98)  # rwidth = relative width of bars
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Distribution of Scores')
plt.show()
```

### Subplots
###### Multiple Charts in one canvas

```python
# plt.subplot(nrows, ncols, index, **kwargs)  # nrows = number of rows, ncols = number of columns, index = index of the subplot

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 35]
y2 = [15, 25, 35, 45, 55]

plt.figure(figsize=(10, 5))  # Set figure size
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.plot(x, y1, marker='o', color='blue', linestyle='--', linewidth=2, markersize=8, label='Data 1')
plt.title('Line Plot 1')

plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
plt.bar(x, y2, color='skyblue', edgecolor='black', width=0.4, label='Data 2')
plt.title('Line Plot 2')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

# fig, ax = plt.subplots(nrows, ncols, figsize=(width, height))  
fig, ax = plt.subplots(2, 2, figsize=(10, 8))  # 2 rows, 2 columns

ax[0, 0].plot(x, y, marker='o', color='blue', linestyle='--', linewidth=2, markersize=8, label='Data 1')
ax[0, 0].set_title('Line Plot ')

ax[1, 0].bar(x, y, color='skyblue', edgecolor='black', width=0.4, label='Data 2')
ax[1, 0].set_title('Bar Chart ')

ax[0, 1].scatter(x, y, color='red', marker='x', s=50, label='Data 3')
ax[0, 1].set_title('Scatter Plot ')

ax[1, 1].hist(y, bins=5, color='green', edgecolor='black', alpha=0.5, rwidth=0.98)
ax[1, 1].set_title('Histogram ')

plt.suptitle('Multiple Charts in One Canvas', fontsize=16)
plt.tight_layout()
plt.show()
```
